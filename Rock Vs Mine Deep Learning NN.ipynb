```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
```

```python
df=pd.read_csv('/media/aj/F/Rock Vs Mine/sonar.all-data.csv',header=None)
```

```python
X=df.drop(60,axis=1)
y=df[60]
```

```python
X
```

```python
y
```

```python
#creating dummie variable
y=pd.get_dummies(y,drop_first=True)
y.sample(5)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1,stratify=y)
```

```python
# Model without drop 
model_without_dropout=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'), # 1st hidden layer
    keras.layers.Dense(30,activation='relu'), # 2nd hidden layer
    keras.layers.Dense(15,activation='relu'), # 3rd hidden layer
    keras.layers.Dense(1,activation='sigmoid')# output layer
])

model_without_dropout.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model_without_dropout.fit(X_train,y_train, epochs=100, batch_size=8)
```

```python
model_without_dropout.evaluate(X_test,y_test) # Evaluate on test set
```

```python
y_predict=model_without_dropout.predict(X_test).reshape(-1) #prediction

y_predict[:10]
```

```python
y_predict= np.round(y_predict)
print(y_predict[:10])
```

```python
np.array(y_test[:10]).reshape(1,-1)
```

```python
np.sum(np.array(y_test[:10]).reshape(1,-1)==y_predict[:10])
```

```python
# printing the metrics
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_predict))
```

```python
# Model with Dropout
model_with_dropout=keras.Sequential([
    keras.layers.Dense(60,input_dim=60, activation='relu'), # 1st hidden layer
    keras.layers.Dropout(0.5),
    
    keras.layers.Dense(30,activation='relu'), # 2nd
    keras.layers.Dropout(0.5),
    
    keras.layers.Dense(15,activation='relu'), # 3rd
    keras.layers.Dropout(0.5),
    
    keras.layers.Dense(1,activation='sigmoid') # output layer
    

])
model_with_dropout.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model_with_dropout.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
model_with_dropout.evaluate(X_train,y_train)
```

```python
model_with_dropout.evaluate(X_test,y_test)
```

```python
y_predict=model_with_dropout.predict(X_test).reshape(-1)
y_predict[:10]
```

```python
y_predict=np.round(y_predict)
y_predict[:10]
```

```python
np.array(y_test[:10]).reshape(1,-1)
```

```python
np.sum(np.array(y_test[:10]).reshape(1,-1)==y_predict[:10])
```

```python
# printing the metrics after dropout
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))
```
